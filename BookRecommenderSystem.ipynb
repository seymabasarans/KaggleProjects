Popularity Based Recommender System

```python
grouped=df.groupby('Book-Title').agg({'Book-Rating':[np.size,np.sum,np.mean]})
```

```python
grouped.head()
```

```python
populer=grouped.sort_values(("Book-Rating","mean"),ascending=False)
```

```python
populer.head()
```

```python
toplam=grouped["Book-Rating"]["sum"].sum()
```

```python
toplam #hepsinin toplamı
```

```python
populer["percentage"]=populer["Book-Rating","sum"].div(toplam)*100
#% lerini hepaspladık
```

```python
populer.head()
```

```python
most_popular=populer.sort_values(("percentage"),ascending=False) 
```
